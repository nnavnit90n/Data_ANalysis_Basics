```python
# Lets come to the name column it requires a little bit more coding

# intialising empty list where we are going to put our names

_names_=[]

#getting name string from the column and appending to the list
for name in a.name:
    _names_.append(name)
#setting a function that will split those name string in  seperate words
def split_name(name):
    spl=str(name).split()
    return spl
#initialising empty list where wwe are going to have words count

_names_for_counts=[]
# getting name string from list and split function later append to the list

for x in _names_:
    for word in split_name(x):
        word=word.lower()
        _names_for_counts.append(word)
```

```python
# we are going to use counter

from collections import Counter
#top 25 word used by hosts

top_25_w=Counter(_names_for_counts).most_common()
top_25_w=top_25_w[0:25]
```

# Using label encode to convert the non numerical labels into the numerical labels

```python
def encode(a):
    for column in a.columns[a.columns.isin(['neighbourhood_group','room_type'])]:
    
        a[column]=a[column].factorize()[0]
    return a

a_1=encode(a.copy())
```

```python
a_1.head(5)
```

```python
corr=a_1.corr(method='kendall')
sns.heatmap(corr,annot=True)
a_1.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import r2_score
```

# Linear Regression::

```python
X=a_1.iloc[:,[0,1,3,4,5]]
print(X)
```

```python
Y=a_1['price'] 
print(Y)
# if we using the Y=a_1['price'].values then we are getting the error that is it has the no header which train test so 
# we have to call it by it simply don't count the values
```

```python
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=1/4,random_state=353)
```

```python
X_train.head()
Y_train.head()
```

```python
x_test.head()
y_test.head()
```

```python
print(X_train.shape)
print(Y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Prepare a linear regressor model
r=LinearRegression()
r.fit(X_train,Y_train)
y_predict=r.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_predict)
```
